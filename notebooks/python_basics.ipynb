## 1️⃣ 变量与数据类型

Python 中的变量不需要声明类型，解释器会自动推断。

```python
student_name = "小明"
age = 20
gpa = 3.75
is_passed = True

print(f"姓名: {student_name}, 类型: {type(student_name)}")
print(f"年龄: {age}, 类型: {type(age)}")
print(f"绩点: {gpa}, 类型: {type(gpa)}")
print(f"是否通过: {is_passed}, 类型: {type(is_passed)}")
```

```python
a = 10
b = 3

print(f"整数除法: {a} // {b} = {a // b}")
print(f"取余: {a} % {b} = {a % b}")
print(f"幂运算: {a} ** {b} = {a ** b}")
print(f"普通除法: {a} / {b} = {a / b}")
```

```python
x = 5
x += 3
print(f"x += 3 后: x = {x}")

x *= 2
print(f"x *= 2 后: x = {x}")

x -= 4
print(f"x -= 4 后: x = {x}")
```

```python
score = 85

print(f"score > 90: {score > 90}")
print(f"score >= 85: {score >= 85}")
print(f"score == 100: {score == 100}")
print(f"score != 60: {score != 60}")

print(f"\n60 < score < 100: {60 < score < 100}")
```

```python
has_homework = True
has_exam = False

print(f"有作业 且 有考试: {has_homework and has_exam}")
print(f"有作业 或 有考试: {has_homework or has_exam}")
print(f"没有作业: {not has_homework}")
```

### 💡 要点

- Python 有四种基本数据类型：`str`（字符串）、`int`（整数）、`float`（浮点数）、`bool`（布尔值）
- `//` 是整数除法，`/` 是普通除法（结果为浮点数）
- Python 支持 `a < x < b` 这样的链式比较
- `and`、`or`、`not` 是逻辑运算符，不是 `&&`、`||`、`!`

## 2️⃣ 字符串操作

字符串是数据处理中最常见的数据类型之一。

```python
name = "Alice"
age = 20
score = 95.678

print(f"f-string: {name}今年{age}岁，成绩{score:.1f}")
print(f"对齐: |{name:>10}|")
print(f"对齐: |{name:<10}|")
print(f"对齐: |{name:^10}|")
print(f"补零: {42:05d}")
```

```python
text = "  Hello, Python!  "

print(f"原始: '{text}'")
print(f"去空格: '{text.strip()}'")
print(f"大写: '{text.upper()}'")
print(f"小写: '{text.lower()}'")
print(f"替换: '{text.replace('Python', 'World')}'")
print(f"分割: {text.strip().split(', ')}")
print(f"查找 'Python' 位置: {text.find('Python')}")
print(f"是否以 'Hello' 开头: {text.strip().startswith('Hello')}")
```

```python
words = ["Python", "机器学习", "数据分析"]
sentence = " | ".join(words)
print(f"join: {sentence}")

csv_line = "小明,85,通过"
parts = csv_line.split(',')
print(f"split: {parts}")
print(f"姓名={parts[0]}, 分数={parts[1]}, 结果={parts[2]}")
```

### 💡 要点

- **f-string** 是 Python 3.6+ 推荐的格式化方式，简洁高效
- `strip()` 去除首尾空白，`split()` 按分隔符拆分，`join()` 按分隔符合并
- 字符串是不可变对象，每次操作都返回新字符串

## 3️⃣ 数据结构

Python 内置了四种常用的数据结构，在数据处理中经常使用。

### 3.1 列表 (List)

列表是最常用的数据结构，可以存储有序的元素集合。

```python
scores = [85, 92, 78, 90, 88]

print(f"列表: {scores}")
print(f"长度: {len(scores)}")
print(f"第一个: {scores[0]}")
print(f"最后一个: {scores[-1]}")
print(f"切片 [1:3]: {scores[1:3]}")
print(f"切片 [:3]: {scores[:3]}")
print(f"切片 [2:]: {scores[2:]}")
```

```python
scores = [85, 92, 78]

scores.append(95)
print(f"append(95): {scores}")

scores.insert(1, 88)
print(f"insert(1, 88): {scores}")

scores.remove(78)
print(f"remove(78): {scores}")

popped = scores.pop()
print(f"pop(): 弹出 {popped}, 剩余 {scores}")

scores.sort()
print(f"sort(): {scores}")

scores.reverse()
print(f"reverse(): {scores}")
```

```python
scores = [85, 92, 78, 90, 88]

print(f"最大值: {max(scores)}")
print(f"最小值: {min(scores)}")
print(f"总和: {sum(scores)}")
print(f"平均值: {sum(scores) / len(scores):.1f}")
print(f"92 在列表中: {92 in scores}")
print(f"100 在列表中: {100 in scores}")
```

### 3.2 元组 (Tuple)

元组与列表类似，但**不可修改**，适合存储不变的数据。

```python
student = ("小明", 20, 3.75)

print(f"元组: {student}")
print(f"姓名: {student[0]}")
print(f"年龄: {student[1]}")

name, age, gpa = student
print(f"解包: name={name}, age={age}, gpa={gpa}")
```

### 3.3 字典 (Dict)

字典以键值对存储数据，查找速度极快，是数据处理的核心数据结构。

```python
student = {
    "name": "小明",
    "age": 20,
    "scores": [85, 92, 78],
    "passed": True
}

print(f"字典: {student}")
print(f"姓名: {student['name']}")
print(f"年龄: {student.get('age')}")
print(f"不存在(默认值): {student.get('rank', '未排名')}")
```

```python
student = {"name": "小明", "age": 20}

student["gpa"] = 3.75
print(f"添加: {student}")

student["age"] = 21
print(f"修改: {student}")

del student["gpa"]
print(f"删除: {student}")

print(f"\n所有键: {list(student.keys())}")
print(f"所有值: {list(student.values())}")
print(f"所有键值对: {list(student.items())}")
```

```python
students = {
    "小明": 85,
    "小红": 92,
    "小刚": 78
}

print("遍历字典:")
for name, score in students.items():
    print(f"  {name}: {score}分")
```

### 3.4 集合 (Set)

集合存储**不重复**的元素，常用于去重和集合运算。

```python
scores = [85, 92, 85, 78, 92, 90]
unique_scores = set(scores)
print(f"原列表: {scores}")
print(f"去重后: {unique_scores}")

a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print(f"\n集合a: {a}")
print(f"集合b: {b}")
print(f"交集: {a & b}")
print(f"并集: {a | b}")
print(f"差集(a-b): {a - b}")
```

### 💡 数据结构选择指南

| 数据结构 | 特点 | 典型用途 |
|---------|------|----------|
| **列表** | 有序、可变 | 存储序列数据、遍历 |
| **元组** | 有序、不可变 | 函数返回多值、固定配置 |
| **字典** | 键值对、查找快 | 数据映射、计数、配置 |
| **集合** | 无序、不重复 | 去重、交集/并集运算 |

## 4️⃣ 控制流

控制流决定程序的执行路径，是编程的基础。

```python
score = 82

if score >= 90:
    grade = "优秀"
elif score >= 80:
    grade = "良好"
elif score >= 70:
    grade = "中等"
elif score >= 60:
    grade = "及格"
else:
    grade = "不及格"

print(f"成绩 {score} 分，等级: {grade}")
```

```python
scores = [85, 92, 78, 90, 88]

print("for 循环遍历:")
for i, score in enumerate(scores):
    print(f"  第{i+1}个学生: {score}分")
```

```python
total = 0
count = 0
n = 1

while n <= 100:
    total += n
    count += 1
    n += 1

print(f"1到100的和: {total}")
print(f"验证: {100 * 101 // 2}")
```

```python
print("break 示例 — 找到第一个不及格的分数:")
scores = [85, 92, 55, 78, 90]
for score in scores:
    if score < 60:
        print(f"  找到: {score}分")
        break

print("\ncontinue 示例 — 只打印及格的分数:")
for score in scores:
    if score < 60:
        continue
    print(f"  {score}分")
```

```python
students = [("小明", 85), ("小红", 92), ("小刚", 78)]

print("同时遍历索引和值:")
for i, (name, score) in enumerate(students):
    print(f"  {i+1}. {name}: {score}分")
```

### 💡 要点

- Python 用**缩进**（通常4个空格）表示代码块，不用大括号
- `elif` 是 `else if` 的缩写
- `enumerate()` 可以同时获取索引和值，非常实用
- `break` 跳出整个循环，`continue` 跳过当前迭代

## 5️⃣ 函数

函数是组织代码的基本单元，让代码可复用、可读、可维护。

```python
def calculate_average(scores):
    """计算平均分"""
    return sum(scores) / len(scores)

scores = [85, 92, 78, 90, 88]
avg = calculate_average(scores)
print(f"平均分: {avg:.2f}")
```

```python
def get_grade(score, passing_line=60):
    """根据分数获取等级，可自定义及格线"""
    if score >= 90:
        return "优秀"
    elif score >= passing_line:
        return "及格"
    else:
        return "不及格"

print(f"85分(默认及格线): {get_grade(85)}")
print(f"55分(默认及格线): {get_grade(55)}")
print(f"55分(及格线70): {get_grade(55, passing_line=70)}")
```

```python
def analyze_scores(*scores, subject="未知科目"):
    """分析成绩，接受任意数量的分数"""
    return {
        "科目": subject,
        "人数": len(scores),
        "最高分": max(scores),
        "最低分": min(scores),
        "平均分": round(sum(scores) / len(scores), 2)
    }

result = analyze_scores(85, 92, 78, 90, 88, subject="数学")
for key, value in result.items():
    print(f"  {key}: {value}")
```

```python
def build_student_info(name, **kwargs):
    """构建学生信息，接受任意关键字参数"""
    info = {"name": name}
    info.update(kwargs)
    return info

student = build_student_info(
    "小明",
    age=20,
    score=85,
    passed=True
)
print(f"学生信息: {student}")
```

### 💡 要点

- 用 `def` 定义函数，用 `return` 返回结果
- 默认参数让函数更灵活：`def f(x, default_val=10)`
- `*args` 接收任意数量的位置参数，`**kwargs` 接收任意数量的关键字参数
- 三引号 `"""..."""` 是文档字符串（docstring），用于说明函数用途

## 6️⃣ 列表推导式

列表推导式是 Python 最优雅的特性之一，用一行代码完成循环和过滤。在数据处理中极为常用。

```python
scores = [85, 92, 78, 55, 90, 45, 88]

passed = [s for s in scores if s >= 60]
print(f"及格的分数: {passed}")

failed = [s for s in scores if s < 60]
print(f"不及格的分数: {failed}")

adjusted = [s + 5 if s < 60 else s for s in scores]
print(f"加分后: {adjusted}")
```

```python
names = ["alice", "BOB", "ChArLiE"]

capitalized = [name.capitalize() for name in names]
print(f"首字母大写: {capitalized}")

name_lengths = {name: len(name) for name in capitalized}
print(f"名字长度: {name_lengths}")
```

```python
students = [
    {"name": "小明", "score": 85},
    {"name": "小红", "score": 92},
    {"name": "小刚", "score": 55},
]

names_passed = [s["name"] for s in students if s["score"] >= 60]
print(f"及格的学生: {names_passed}")

score_map = {s["name"]: s["score"] for s in students}
print(f"姓名-分数映射: {score_map}")
```

### 💡 要点

- 基本语法：`[表达式 for 变量 in 可迭代对象 if 条件]`
- 字典推导式：`{键表达式: 值表达式 for 变量 in 可迭代对象}`
- 列表推导式比等价的 for 循环更简洁、执行更快
- 如果逻辑过于复杂，建议改用普通 for 循环以保持可读性

## 7️⃣ 文件读写

数据科学中经常需要读取和写入文件。

```python
with open('sample_scores.csv', 'r', encoding='utf-8') as f:
    lines = f.readlines()

header = lines[0].strip().split(',')
print(f"表头: {header}")

for line in lines[1:]:
    parts = line.strip().split(',')
    print(f"  学号={parts[0]}, 分数={parts[1]}, 等级={parts[2]}")
```

## 8️⃣ 常用内置函数与模块导入

Python 提供了丰富的内置函数和标准库，也是使用第三方库的基础。

```python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

print(f"排序: {sorted(numbers)}")
print(f"逆序排序: {sorted(numbers, reverse=True)}")
print(f"枚举: {list(enumerate(['a', 'b', 'c']))}")
print(f"zip: {list(zip(['a', 'b', 'c'], [1, 2, 3]))}")
print(f"map: {list(map(str, [1, 2, 3]))}")
print(f"filter: {list(filter(lambda x: x > 3, numbers))}")
```

```python
import math

print(f"圆周率: {math.pi:.6f}")
print(f"平方根: {math.sqrt(144)}")
print(f"向上取整: {math.ceil(3.2)}")
print(f"向下取整: {math.floor(3.8)}")
```

```python
from collections import Counter

grades = ["优秀", "良好", "及格", "良好", "优秀", "优秀", "不及格"]
counter = Counter(grades)

print(f"等级统计: {dict(counter)}")
print(f"最常见的2个: {counter.most_common(2)}")
```

```python
import numpy as np

arr = np.array([85, 92, 78, 90, 88])
print(f"NumPy数组: {arr}")
print(f"均值: {arr.mean():.2f}")
print(f"标准差: {arr.std():.2f}")
print(f"最大值索引: {arr.argmax()}")

arr_scaled = (arr - arr.mean()) / arr.std()
print(f"标准化后: {arr_scaled.round(2)}")
```

### 💡 要点

- `import 模块名` 导入整个模块，`from 模块名 import 函数名` 导入特定函数
- `sorted()`、`enumerate()`、`zip()`、`map()`、`filter()` 是最常用的高阶内置函数
- `collections.Counter` 在统计频次时非常方便
- NumPy 是数据科学的基础库，后续 ML 教程会大量使用

## 9️⃣ 综合练习

用所学知识完成以下练习，为后续机器学习教程做好准备。

```python
students = [
    {"name": "小明", "math": 85, "english": 78, "science": 92},
    {"name": "小红", "math": 92, "english": 88, "science": 85},
    {"name": "小刚", "math": 55, "english": 62, "science": 58},
    {"name": "小丽", "math": 78, "english": 95, "science": 82},
    {"name": "小强", "math": 45, "english": 52, "science": 48},
]

for s in students:
    scores = [s["math"], s["english"], s["science"]]
    avg = sum(scores) / len(scores)
    s["average"] = round(avg, 2)
    s["passed"] = avg >= 60

print("=== 学生成绩分析 ===")
for s in students:
    status = "✅ 通过" if s["passed"] else "❌ 未通过"
    print(f"  {s['name']}: 均分={s['average']}, {status}")
```

```python
passed_students = [s for s in students if s["passed"]]
failed_students = [s for s in students if not s["passed"]]

print(f"通过人数: {len(passed_students)}")
print(f"未通过人数: {len(failed_students)}")
print(f"通过率: {len(passed_students) / len(students) * 100:.1f}%")

all_averages = [s["average"] for s in students]
print(f"\n班级均分: {sum(all_averages) / len(all_averages):.2f}")
print(f"最高均分: {max(all_averages):.2f}")
print(f"最低均分: {min(all_averages):.2f}")
```

```python
subject_scores = {
    "数学": [s["math"] for s in students],
    "英语": [s["english"] for s in students],
    "科学": [s["science"] for s in students],
}

print("=== 各科目统计 ===")
for subject, scores in subject_scores.items():
    avg = sum(scores) / len(scores)
    print(f"  {subject}: 均分={avg:.1f}, 最高={max(scores)}, 最低={min(scores)}")
```
