## 1️⃣ 变量与数据类型

Python 中的变量不需要声明类型，解释器会自动推断。

```python
patient_name = "张三"
age = 55
bmi = 26.3
is_hypertensive = True

print(f"姓名: {patient_name}, 类型: {type(patient_name)}")
print(f"年龄: {age}, 类型: {type(age)}")
print(f"BMI: {bmi}, 类型: {type(bmi)}")
print(f"是否高血压: {is_hypertensive}, 类型: {type(is_hypertensive)}")
```

```python
a = 145
b = 90

print(f"收缩压/舒张压: {a}/{b}")
print(f"脉压差: {a - b}")
print(f"平均动脉压: {a // 3 + b * 2 // 3}")
print(f"血压比值: {a / b:.2f}")
```

```python
glucose = 110
glucose += 15
print(f"餐后血糖: {glucose}")

glucose *= 0.9
print(f"运动后血糖: {glucose:.1f}")

glucose -= 5
print(f"用药后血糖: {glucose:.1f}")
```

```python
bp_systolic = 135

print(f"收缩压 > 140: {bp_systolic > 140}")
print(f"收缩压 >= 135: {bp_systolic >= 135}")
print(f"收缩压 == 120: {bp_systolic == 120}")
print(f"收缩压 != 120: {bp_systolic != 120}")

print(f"\n90 < 收缩压 < 140: {90 < bp_systolic < 140}")
```

```python
has_hypertension = True
has_diabetes = False

print(f"高血压 且 糖尿病: {has_hypertension and has_diabetes}")
print(f"高血压 或 糖尿病: {has_hypertension or has_diabetes}")
print(f"没有高血压: {not has_hypertension}")
```

### 💡 要点

- Python 有四种基本数据类型：`str`（字符串）、`int`（整数）、`float`（浮点数）、`bool`（布尔值）
- `//` 是整数除法，`/` 是普通除法（结果为浮点数）
- Python 支持 `a < x < b` 这样的链式比较
- `and`、`or`、`not` 是逻辑运算符，不是 `&&`、`||`、`!`

## 2️⃣ 字符串操作

字符串是数据处理中最常见的数据类型之一。

```python
patient_id = "PT0042"
age = 55
glucose = 126.578

print(f"f-string: 患者{patient_id}，年龄{age}岁，血糖{glucose:.1f}")
print(f"对齐: |{patient_id:>10}|")
print(f"对齐: |{patient_id:<10}|")
print(f"对齐: |{patient_id:^10}|")
print(f"补零: {42:05d}")
```

```python
text = "  Hypertension, Diabetes  "

print(f"原始: '{text}'")
print(f"去空格: '{text.strip()}'")
print(f"大写: '{text.upper()}'")
print(f"小写: '{text.lower()}'")
print(f"替换: '{text.replace('Diabetes', 'Hyperlipidemia')}'")
print(f"分割: {text.strip().split(', ')}")
print(f"查找 'Diabetes' 位置: {text.find('Diabetes')}")
print(f"是否以 'Hypertension' 开头: {text.strip().startswith('Hypertension')}")
```

```python
diagnoses = ["高血压", "糖尿病", "高血脂"]
combined = " | ".join(diagnoses)
print(f"join: {combined}")

csv_line = "PT0042,55,26.3,患病"
parts = csv_line.split(',')
print(f"split: {parts}")
print(f"患者ID={parts[0]}, 年龄={parts[1]}, BMI={parts[2]}, 诊断={parts[3]}")
```

### 💡 要点

- **f-string** 是 Python 3.6+ 推荐的格式化方式，简洁高效
- `strip()` 去除首尾空白，`split()` 按分隔符拆分，`join()` 按分隔符合并
- 字符串是不可变对象，每次操作都返回新字符串

## 3️⃣ 数据结构

Python 内置了四种常用的数据结构，在数据处理中经常使用。

### 3.1 列表 (List)

列表是最常用的数据结构，可以存储有序的元素集合。

```python
glucose_readings = [95, 110, 126, 88, 102]

print(f"血糖读数: {glucose_readings}")
print(f"长度: {len(glucose_readings)}")
print(f"第一次: {glucose_readings[0]}")
print(f"最后一次: {glucose_readings[-1]}")
print(f"切片 [1:3]: {glucose_readings[1:3]}")
print(f"切片 [:3]: {glucose_readings[:3]}")
print(f"切片 [2:]: {glucose_readings[2:]}")
```

```python
glucose_readings = [95, 110, 126]

glucose_readings.append(135)
print(f"append(135): {glucose_readings}")

glucose_readings.insert(1, 105)
print(f"insert(1, 105): {glucose_readings}")

glucose_readings.remove(126)
print(f"remove(126): {glucose_readings}")

popped = glucose_readings.pop()
print(f"pop(): 弹出 {popped}, 剩余 {glucose_readings}")

glucose_readings.sort()
print(f"sort(): {glucose_readings}")

glucose_readings.reverse()
print(f"reverse(): {glucose_readings}")
```

```python
glucose_readings = [95, 110, 126, 88, 102]

print(f"最高值: {max(glucose_readings)}")
print(f"最低值: {min(glucose_readings)}")
print(f"总和: {sum(glucose_readings)}")
print(f"平均值: {sum(glucose_readings) / len(glucose_readings):.1f}")
print(f"126 在列表中: {126 in glucose_readings}")
print(f"200 在列表中: {200 in glucose_readings}")
```

### 3.2 元组 (Tuple)

元组与列表类似，但**不可修改**，适合存储不变的数据。

```python
vital_signs = ("张三", 135, 85, 72)

print(f"元组: {vital_signs}")
print(f"姓名: {vital_signs[0]}")
print(f"收缩压: {vital_signs[1]}")

name, systolic, diastolic, heart_rate = vital_signs
print(f"解包: name={name}, 收缩压={systolic}, 舒张压={diastolic}, 心率={heart_rate}")
```

### 3.3 字典 (Dict)

字典以键值对存储数据，查找速度极快，是数据处理的核心数据结构。

```python
patient = {
    "name": "张三",
    "age": 55,
    "vitals": [135, 85, 72],
    "diagnosed": True
}

print(f"字典: {patient}")
print(f"姓名: {patient['name']}")
print(f"年龄: {patient.get('age')}")
print(f"不存在(默认值): {patient.get('ward', '未分配')}")
```

```python
patient = {"name": "张三", "age": 55}

patient["bmi"] = 26.3
print(f"添加: {patient}")

patient["age"] = 56
print(f"修改: {patient}")

del patient["bmi"]
print(f"删除: {patient}")

print(f"\n所有键: {list(patient.keys())}")
print(f"所有值: {list(patient.values())}")
print(f"所有键值对: {list(patient.items())}")
```

```python
patients = {
    "张三": 135,
    "李四": 122,
    "王五": 148
}

print("遍历字典:")
for name, bp in patients.items():
    status = "偏高" if bp > 140 else "正常"
    print(f"  {name}: 收缩压={bp}mmHg ({status})")
```

### 3.4 集合 (Set)

集合存储**不重复**的元素，常用于去重和集合运算。

```python
diagnoses_list = ["高血压", "糖尿病", "高血压", "高血脂", "糖尿病", "冠心病"]
unique_diagnoses = set(diagnoses_list)
print(f"原列表: {diagnoses_list}")
print(f"去重后: {unique_diagnoses}")

patient_a_diagnoses = {"高血压", "糖尿病", "高血脂"}
patient_b_diagnoses = {"糖尿病", "冠心病", "高血脂"}
print(f"\n患者A: {patient_a_diagnoses}")
print(f"患者B: {patient_b_diagnoses}")
print(f"共有疾病(交集): {patient_a_diagnoses & patient_b_diagnoses}")
print(f"所有疾病(并集): {patient_a_diagnoses | patient_b_diagnoses}")
print(f"A独有(A-B): {patient_a_diagnoses - patient_b_diagnoses}")
```

### 💡 数据结构选择指南

| 数据结构 | 特点 | 典型用途 |
|---------|------|----------|
| **列表** | 有序、可变 | 存储检验指标序列、时间序列数据 |
| **元组** | 有序、不可变 | 函数返回多值、固定配置 |
| **字典** | 键值对、查找快 | 患者信息映射、指标编码、配置 |
| **集合** | 无序、不重复 | 诊断去重、共有疾病计算 |

## 4️⃣ 控制流

控制流决定程序的执行路径，是编程的基础。

```python
bp_systolic = 135

if bp_systolic >= 140:
    bp_level = "高血压"
elif bp_systolic >= 130:
    bp_level = "偏高"
elif bp_systolic >= 120:
    bp_level = "正常高值"
elif bp_systolic >= 90:
    bp_level = "正常"
else:
    bp_level = "偏低"

print(f"收缩压 {bp_systolic} mmHg，等级: {bp_level}")
```

```python
glucose_readings = [95, 110, 126, 88, 102]

print("for 循环遍历:")
for i, glucose in enumerate(glucose_readings):
    status = "偏高" if glucose > 100 else "正常"
    print(f"  第{i+1}次测量: {glucose} mg/dL ({status})")
```

```python
total = 0
count = 0
n = 1

while n <= 100:
    total += n
    count += 1
    n += 1

print(f"1到100的和: {total}")
print(f"验证: {100 * 101 // 2}")
```

```python
print("break 示例 — 找到第一个血糖异常的读数:")
glucose_readings = [95, 110, 135, 88, 102]
for glucose in glucose_readings:
    if glucose > 126:
        print(f"  找到异常: {glucose} mg/dL")
        break

print("\ncontinue 示例 — 只打印正常血糖:")
for glucose in glucose_readings:
    if glucose > 126:
        continue
    print(f"  {glucose} mg/dL")
```

```python
patients = [("张三", 135), ("李四", 122), ("王五", 148)]

print("同时遍历索引和值:")
for i, (name, bp) in enumerate(patients):
    print(f"  {i+1}. {name}: 收缩压={bp}mmHg")
```

### 💡 要点

- Python 用**缩进**（通常4个空格）表示代码块，不用大括号
- `elif` 是 `else if` 的缩写
- `enumerate()` 可以同时获取索引和值，非常实用
- `break` 跳出整个循环，`continue` 跳过当前迭代

## 5️⃣ 函数

函数是组织代码的基本单元，让代码可复用、可读、可维护。

```python
def calculate_average(values):
    """计算平均值"""
    return sum(values) / len(values)

glucose_readings = [95, 110, 126, 88, 102]
avg = calculate_average(glucose_readings)
print(f"平均血糖: {avg:.2f} mg/dL")
```

```python
def classify_bmi(bmi, standard="WHO"):
    """根据BMI分类，可切换标准"""
    if standard == "WHO":
        if bmi < 18.5:
            return "偏瘦"
        elif bmi < 25:
            return "正常"
        elif bmi < 30:
            return "超重"
        else:
            return "肥胖"
    else:
        if bmi < 18.5:
            return "偏瘦"
        elif bmi < 24:
            return "正常"
        elif bmi < 28:
            return "超重"
        else:
            return "肥胖"

print(f"BMI=26.3 (WHO标准): {classify_bmi(26.3)}")
print(f"BMI=26.3 (中国标准): {classify_bmi(26.3, standard='中国')}")
```

```python
def analyze_vitals(*values, patient_name="未知患者"):
    """分析生命体征，接受任意数量的测量值"""
    return {
        "患者": patient_name,
        "测量次数": len(values),
        "最高值": max(values),
        "最低值": min(values),
        "平均值": round(sum(values) / len(values), 2)
    }

result = analyze_vitals(135, 128, 142, 130, 138, patient_name="张三")
for key, value in result.items():
    print(f"  {key}: {value}")
```

```python
def build_patient_record(name, **kwargs):
    """构建患者记录，接受任意关键字参数"""
    record = {"name": name}
    record.update(kwargs)
    return record

patient = build_patient_record(
    "张三",
    age=55,
    bmi=26.3,
    diagnosed=True
)
print(f"患者记录: {patient}")
```

### 💡 要点

- 用 `def` 定义函数，用 `return` 返回结果
- 默认参数让函数更灵活：`def f(x, default_val=10)`
- `*args` 接收任意数量的位置参数，`**kwargs` 接收任意数量的关键字参数
- 三引号 `"""..."""` 是文档字符串（docstring），用于说明函数用途

## 6️⃣ 列表推导式

列表推导式是 Python 最优雅的特性之一，用一行代码完成循环和过滤。在数据处理中极为常用。

```python
glucose_readings = [95, 110, 126, 88, 102, 135, 78]

normal = [g for g in glucose_readings if g <= 100]
print(f"正常血糖(≤100): {normal}")

elevated = [g for g in glucose_readings if g > 100]
print(f"偏高血糖(>100): {elevated}")

adjusted = [g * 0.9 if g > 126 else g for g in glucose_readings]
print(f"治疗后(>126降10%): {adjusted}")
```

```python
diagnoses = ["hypertension", "DIABETES", "Hyperlipidemia"]

capitalized = [d.capitalize() for d in diagnoses]
print(f"首字母大写: {capitalized}")

diag_lengths = {d: len(d) for d in capitalized}
print(f"诊断名长度: {diag_lengths}")
```

```python
patients = [
    {"name": "张三", "glucose": 135},
    {"name": "李四", "glucose": 95},
    {"name": "王五", "glucose": 148},
]

abnormal_names = [p["name"] for p in patients if p["glucose"] > 126]
print(f"血糖异常的患者: {abnormal_names}")

glucose_map = {p["name"]: p["glucose"] for p in patients}
print(f"姓名-血糖映射: {glucose_map}")
```

### 💡 要点

- 基本语法：`[表达式 for 变量 in 可迭代对象 if 条件]`
- 字典推导式：`{键表达式: 值表达式 for 变量 in 可迭代对象}`
- 列表推导式比等价的 for 循环更简洁、执行更快
- 如果逻辑过于复杂，建议改用普通 for 循环以保持可读性

## 8️⃣ 常用内置函数与模块导入

Python 提供了丰富的内置函数和标准库，也是使用第三方库的基础。

```python
glucose_values = [95, 110, 126, 88, 102, 135, 78]

print(f"排序: {sorted(glucose_values)}")
print(f"逆序排序: {sorted(glucose_values, reverse=True)}")
print(f"枚举: {list(enumerate(['收缩压', '舒张压', '心率']))}")
print(f"zip: {list(zip(['收缩压', '舒张压', '心率'], [135, 85, 72]))}")
print(f"map: {list(map(str, [135, 85, 72]))}")
print(f"filter: {list(filter(lambda x: x > 100, glucose_values))}")
```

```python
import math

print(f"圆周率: {math.pi:.6f}")
print(f"平方根: {math.sqrt(144)}")
print(f"向上取整: {math.ceil(3.2)}")
print(f"向下取整: {math.floor(3.8)}")
```

```python
from collections import Counter

diagnoses = ["高血压", "糖尿病", "高血脂", "糖尿病", "高血压", "高血压", "冠心病"]
counter = Counter(diagnoses)

print(f"诊断统计: {dict(counter)}")
print(f"最常见的2个: {counter.most_common(2)}")
```

```python
import numpy as np

glucose_arr = np.array([95, 110, 126, 88, 102])
print(f"NumPy数组: {glucose_arr}")
print(f"均值: {glucose_arr.mean():.2f}")
print(f"标准差: {glucose_arr.std():.2f}")
print(f"最大值索引: {glucose_arr.argmax()}")

glucose_scaled = (glucose_arr - glucose_arr.mean()) / glucose_arr.std()
print(f"标准化后: {glucose_scaled.round(2)}")
```

### 💡 要点

- `import 模块名` 导入整个模块，`from 模块名 import 函数名` 导入特定函数
- `sorted()`、`enumerate()`、`zip()`、`map()`、`filter()` 是最常用的高阶内置函数
- `collections.Counter` 在统计频次时非常方便
- NumPy 是数据科学的基础库，后续 ML 教程会大量使用

## 9️⃣ 综合练习

用所学知识完成以下练习，为后续机器学习教程做好准备。

```python
patients = [
    {"name": "张三", "glucose": 135, "bp": 148, "cholesterol": 6.8},
    {"name": "李四", "glucose": 95, "bp": 122, "cholesterol": 4.5},
    {"name": "王五", "glucose": 148, "bp": 155, "cholesterol": 7.2},
    {"name": "赵六", "glucose": 88, "bp": 118, "cholesterol": 4.2},
    {"name": "钱七", "glucose": 165, "bp": 160, "cholesterol": 8.1},
]

for p in patients:
    risk_score = 0
    if p["glucose"] > 126: risk_score += 1
    if p["bp"] > 140: risk_score += 1
    if p["cholesterol"] > 5.2: risk_score += 1
    p["risk_score"] = risk_score
    p["diagnosed"] = risk_score >= 2

print("=== 患者风险评估 ===")
for p in patients:
    status = "⚠️ 高风险" if p["diagnosed"] else "✅ 低风险"
    print(f"  {p['name']}: 风险分={p['risk_score']}, {status}")
```

```python
high_risk = [p for p in patients if p["diagnosed"]]
low_risk = [p for p in patients if not p["diagnosed"]]

print(f"高风险人数: {len(high_risk)}")
print(f"低风险人数: {len(low_risk)}")
print(f"高风险比例: {len(high_risk) / len(patients) * 100:.1f}%")

all_glucose = [p["glucose"] for p in patients]
print(f"\n平均血糖: {sum(all_glucose) / len(all_glucose):.1f}")
print(f"最高血糖: {max(all_glucose)}")
print(f"最低血糖: {min(all_glucose)}")
```

```python
indicators = {
    "血糖": [p["glucose"] for p in patients],
    "收缩压": [p["bp"] for p in patients],
    "胆固醇": [p["cholesterol"] for p in patients],
}

print("=== 各指标统计 ===")
for indicator, values in indicators.items():
    avg = sum(values) / len(values)
    print(f"  {indicator}: 均值={avg:.1f}, 最高={max(values)}, 最低={min(values)}")
```
